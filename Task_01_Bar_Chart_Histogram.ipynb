```python

df = pd.read_csv(r'D:\Data Science Internship (prodigy infotech)\New folder\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)

print(df.shape)
print(df.columns[:10])
df.head()
```

```python
top10 = df['Country Name'].value_counts().head(10).index
filtered_df = df[df['Country Name'].isin(top10)]

plt.figure(figsize=(10, 6))
sns.countplot(x='Country Name', data=filtered_df, palette='Set2')
plt.title('Top 10 Countries by Record Count')
plt.xlabel('Country')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
pop_2020 = df['2020'].dropna()

plt.figure(figsize=(10, 6))
sns.histplot(pop_2020, bins=30, kde=True, color='skyblue')
plt.title('Distribution of Population Across Countries (2020)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.tight_layout()
plt.show()
```
